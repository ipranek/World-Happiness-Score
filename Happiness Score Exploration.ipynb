```python
mean_life_ladder = df.groupby("Country name")["Life Ladder"].mean()
lowest_life_ladder_country = mean_life_ladder.idxmin()
lowest_life_ladder_value = mean_life_ladder.min()

sorted_mean_life_ladder = mean_life_ladder.sort_values(ascending= False)
highest_life_ladder_country = sorted_mean_life_ladder.head(1)
print(f"The country with the highest life ladder is {highest_life_ladder_country.index[0]} and its mean life ladder score is {highest_life_ladder_country.values[0]:.2f}.")

print(f"The country with the lowest life ladder is {lowest_life_ladder_country} and its mean life ladder score is {lowest_life_ladder_value}.")
```
